```python
import pandas as pd
import pycountry_convert as pc

df = pd.read_csv("data\energy-pop-exposure-nuclear-plants-locations_plants.csv")

```

```python
countries = df["Country"]
num_reactor = df["NumReactor"]
res_lst = []

for i in range(len(countries)):
    curr_country1 = countries[i]
    curr_num_reactor1 = num_reactor[i]
    curr_country_reactor_sum = curr_num_reactor1

    for j in range(i + 1,len(countries)):
        curr_country2 = countries[j]
        curr_num_reactor2 = num_reactor[j]
    
        if curr_country1 == curr_country2:
            curr_country_reactor_sum += curr_num_reactor2

    res_lst.append((i, curr_country_reactor_sum, curr_country1))

print(countries)
res_lst



```

```python
i = 0

res_lst

while i < len(res_lst):
    curr_tuple1 = res_lst[i]
    print("i" + str(i), curr_tuple1)
    j = i + 1
    while j < len(res_lst):
        curr_tuple2 = res_lst[j]
        print("j" + str(j), curr_tuple2)
        if (isinstance(curr_tuple1[2], str) and isinstance(curr_tuple2[2], str)) == False:
            raise Exception

        if str(curr_tuple1[2]) == str(curr_tuple2[2]):
            res_lst.pop(j)
        j += 1
    i += 1

res_lst
```

```python
df2 = pd.DataFrame(res_lst, columns =['index', 'nuclear_reactor_freq', 'country'])



df2.to_csv("country_and_reactor_freq.csv")
```
