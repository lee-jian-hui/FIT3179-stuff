```python
import pandas as pd
import pycountry_convert as pc

df = pd.read_csv("data\energy-pop-exposure-nuclear-plants-locations_plants.csv")

```

```python
countries = df["Country"]
num_reactor = df["NumReactor"]
res_lst = []

for i in range(len(countries)):
    curr_country1 = countries[i]
    curr_num_reactor1 = num_reactor[i]
    curr_country_reactor_sum = curr_num_reactor1

    for j in range(i + 1,len(countries)):
        curr_country2 = countries[j]
        curr_num_reactor2 = num_reactor[j]
    
        if curr_country1 == curr_country2:
            curr_country_reactor_sum += curr_num_reactor2

    res_lst.append((i, curr_country_reactor_sum, curr_country1))

print(countries)
res_lst



```
